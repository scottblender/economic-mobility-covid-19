```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os  
```

```python
os.getcwd()
```

```python
mobilitydata = pd.read_csv("C:\\Users\\Scott Blender\\Downloads\\Region_Mobility_Report_CSVs\\2020_US_Region_Mobility_Report.csv")
```

```python
mobilitydata = mobilitydata[(mobilitydata['sub_region_1']=='Pennsylvania') & (mobilitydata['sub_region_2']=='Montgomery County')]
```

```python
mobilitydata = mobilitydata.fillna((mobilitydata['retail_and_recreation_percent_change_from_baseline']+mobilitydata['grocery_and_pharmacy_percent_change_from_baseline']+mobilitydata['parks_percent_change_from_baseline']+mobilitydata['transit_stations_percent_change_from_baseline']+mobilitydata['workplaces_percent_change_from_baseline']+mobilitydata['residential_percent_change_from_baseline']).mean())
mobilitydata.head()
```

```python
df_date = mobilitydata['date']
df_retail = mobilitydata['retail_and_recreation_percent_change_from_baseline']
df_grocery = mobilitydata['grocery_and_pharmacy_percent_change_from_baseline']
df_parks = mobilitydata['parks_percent_change_from_baseline']
df_transit = mobilitydata['transit_stations_percent_change_from_baseline']
df_workplaces = mobilitydata['workplaces_percent_change_from_baseline']
df_residential = mobilitydata['residential_percent_change_from_baseline']
```

```python
mpl.rcParams['figure.dpi'] = 200
# create a figure and axis
fig, ax = plt.subplots()
#plot for Retail Mobility
ax.plot(df_date, df_retail, label = 'Retail')
#plot for Grocery Mobility
ax.plot(df_date, df_grocery, label = 'Grocery')
#plot for Parks Mobility
ax.plot(df_date, df_parks, label = 'Parks')
#plot for Transit Mobility
ax.plot(df_date, df_transit, label = 'Transit')
#plot for Wokplaces Mobility
ax.plot(df_date, df_workplaces, label = 'Workplace')
#plot for Residential Mobility
ax.plot(df_date, df_residential, label = 'Residential')
# set a title and labels
ax.set_xlabel('Dates')
ax.legend(prop={'size': 5})
ax.set_ylabel("% Change in Mobility")
ax.set_title('Montgomery County Mobility Changes since February 15th, 2020')
plt.xticks(np.arange(0, len(df_date)+1, 70))
plt.show()
```
